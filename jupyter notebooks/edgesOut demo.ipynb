```python
import pandas as pandas
from biothings_explorer.query.predict import Predict
from biothings_explorer.query.visualize import display_graph
from biothings_explorer.user_query_dispatcher import FindConnection
from biothings_explorer.hint import Hint
import nest_asyncio
nest_asyncio.apply()
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python

disease_name = 'COVID-19'
ht = Hint()
disease = ht.query(disease_name)['Disease'][0]
disease
```

```python
direct_gene_dict = {'TNF': 4, 'IL2': 1, 'IL1B': 1, 'IFNA1': 1, 'TNNI3': 1, 'IL6': 1}
```

```python
gene_inputs = []
for gene in list(direct_gene_dict.keys()): 
    try: 
        gene_input = ht.query(gene)["Gene"][0]
        gene_inputs.append(gene_input)
    except: 
        print(gene + ' Failed')
```

```python
gene_inputs
```

```python

pd3 = Predict(
    input_objs=gene_inputs,
    intermediate_nodes=[['Gene', 'Disease', 'PhenotypicFeature']], 
    output_types =['Pathway'], 
    config={"expand": False, "annotate": ["nodeDegree", "edgesOut"]}
)
pd3.connect(verbose=True)
```

```python
pd3.display_table_view(extra_fields=["edgesOut"])
```
