```python
from biothings_explorer.query.predict import Predict
from biothings_explorer.query.visualize import display_graph
from biothings_explorer.hint import Hint
import nest_asyncio
nest_asyncio.apply()
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python

ht = Hint()
# find all potential representations of hyperphenylalaninemia
pcd = ht.query("MONDO:0016575")
# select the correct representation of hyperphenylalaninemia
pcd = pcd['Disease'][0]
pcd
```

```python
pd = Predict(
    input_objs=[pcd],
    intermediate_nodes =['Gene'], 
    output_types =['Pathway'], 
    config={"expand": True, "annotate": ["nodeDegree", "edgesOut"]}
)
pd.connect(verbose=True)
```

```python
pd.display_table_view(extra_fields=["edgesOut"]).sort_values(by="pred2_edgesOut")
```
